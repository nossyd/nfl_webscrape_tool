```python
### imports
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
### Gather URL
URL = 'https://www.pro-football-reference.com/play-index/pgl_finder.cgi?request=1&match=game&year_min=2019&year_max=2019&season_start=1&season_end=-1&age_min=0&age_max=99&game_type=A&league_id=&team_id=&opp_id=&game_num_min=0&game_num_max=99&week_num_min=1&week_num_max=1&game_day_of_week=&game_location=&game_result=&handedness=&is_active=&is_hof=&c1stat=pass_att&c1comp=gt&c1val=1&c2stat=&c2comp=gt&c2val=&c3stat=&c3comp=gt&c3val=&c4stat=&c4comp=gt&c4val=&order_by=pass_rating&from_link=1'
response = requests.get(URL)
html = response.text

print(html)
```

```python
### Parse HTML Contents
soup = BeautifulSoup(html, "html.parser")
```

```python
### Set rows into list of lists
nfl_2019_info_raw = []

for tr in soup.find_all('tr'):
    cols = []
    for td in tr.find_all(['td', 'th']):
        td_text = td.get_text(strip=True)
        if len(td_text):
            cols.append(td_text)
    nfl_2019_info_raw.append(cols)

print(nfl_2019_info_raw)
```

```python
### Delete first + duplicate list
del nfl_2019_info_raw[0]
del nfl_2019_info_raw[21]

print(nfl_2019_info_raw)
```

```python
### Delete '@'
#nfl_2019_info_raw = [x for x in nfl_2019_info_raw if x != '@']

# Remove "@"
# Removing element from list of lists
N = '@'
nfl_2019_info_raw = [[ele for ele in sub if ele != N] for sub in nfl_2019_info_raw] 
```

```python
### Unzip list
nfl_2019_qb_wk1_rank, nfl_2019_qb_wk1_player, nfl_2019_qb_wk1_pos, nfl_2019_qb_wk1_age, nfl_2019_qb_wk1_date, nfl_2019_qb_wk1_lgthrw, nfl_2019_qb_wk1_tm, nfl_2019_qb_wk1_opp, nfl_2019_qb_wk1_result, nfl_2019_qb_wk1_game, nfl_2019_qb_wk1_week, nfl_2019_qb_wk1_day, nfl_2019_qb_wk1_cmp, nfl_2019_qb_wk1_att, nfl_2019_qb_wk1_cmpper, nfl_2019_qb_wk1_yds, nfl_2019_qb_wk1_td, nfl_2019_qb_wk1_int, nfl_2019_qb_wk1_rate, nfl_2019_qb_wk1_sk, nfl_2019_qb_wk1_ydslost, nfl_2019_qb_wk1_ya, nfl_2019_qb_wk1_aya  = map(list, zip(*nfl_2019_info_raw))
```

```python
### Confirm split + is a list
print(nfl_2019_qb_wk1_rank)
type(nfl_2019_qb_wk1_rank)

print(nfl_2019_qb_wk1_player)
type(nfl_2019_qb_wk1_player)
```

```python
### Merge into df
nfl_2019_qb_wk1_df = pd.DataFrame(
    {'Rank': nfl_2019_qb_wk1_rank,
     'Player': nfl_2019_qb_wk1_player,
     'Position': nfl_2019_qb_wk1_pos, 
     'Age': nfl_2019_qb_wk1_age, 
     'Date': nfl_2019_qb_wk1_date, 
     'League': nfl_2019_qb_wk1_lgthrw, 
     'Team': nfl_2019_qb_wk1_tm, 
     'Opp': nfl_2019_qb_wk1_opp, 
     'Result': nfl_2019_qb_wk1_result, 
     'Game': nfl_2019_qb_wk1_game, 
     'Week': nfl_2019_qb_wk1_week, 
     'Day': nfl_2019_qb_wk1_day, 
     'Cmp': nfl_2019_qb_wk1_cmp, 
     'Pass_Att': nfl_2019_qb_wk1_att, 
     'Cmp_Per': nfl_2019_qb_wk1_cmpper, 
     'Pass_Yards': nfl_2019_qb_wk1_yds, 
     'Pass_TD': nfl_2019_qb_wk1_td, 
     'Pass_Int': nfl_2019_qb_wk1_int, 
     'QB_Rate': nfl_2019_qb_wk1_rate, 
     'Sacked': nfl_2019_qb_wk1_sk, 
     'Yds_Lost': nfl_2019_qb_wk1_ydslost, 
     'Yards_Att': nfl_2019_qb_wk1_ya, 
     'Adj_Yards_Att': nfl_2019_qb_wk1_aya
    })


print(nfl_2019_qb_wk1_df)
```

```python
nfl_2019_qb_wk1_df.columns = nfl_2019_qb_wk1_df.iloc[0]
nfl_2019_qb_wk1_df = nfl_2019_qb_wk1_df.reindex(nfl_2019_qb_wk1_df.index.drop(0)).reset_index(drop=True)
nfl_2019_qb_wk1_df.columns.name = None
print(nfl_2019_qb_wk1_df)
```
